# TOPIC

Extracción de los tópicos de submuestras de los textos.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import NMF,LatentDirichletAllocation
from spacy.lang.es.stop_words import STOP_WORDS
```

```python
#Parámetros
n_samples = 30000
n_features = 2000
n_components = 10
n_top_words = 20
stop_words=list(STOP_WORDS)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic #%d: " % topic_idx
        message += ",".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
```

```python
DF1.head()
#DF2.head()
#DF3.head()
```

```python
#Submuestra
data_sample=DF1.sample(n=n_samples)

data_sample.head()
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2,
                                   max_features=n_features,
                                   stop_words=stop_words,ngram_range=(2, 2))
```

```python
tfidf = tfidf_vectorizer.fit_transform(data_sample)
```

```python
nmf = NMF(n_components=n_components, random_state=1,
          alpha=.1, l1_ratio=.5).fit(tfidf)
```

```python
print("\nTopics in NMF model (Frobenius norm):")
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
print_top_words(nmf, tfidf_feature_names, n_top_words)
```

```python
print("Fitting the NMF model (generalized Kullback-Leibler divergence) with "
      "tf-idf features, n_samples=%d and n_features=%d..."
      % (n_samples, n_features))

nmf = NMF(n_components=n_components, random_state=1,
          beta_loss='kullback-leibler', solver='mu', max_iter=1000, alpha=.1,
          l1_ratio=.5).fit(tfidf)
```

```python
print("\nTopics in NMF model (generalized Kullback-Leibler divergence):")
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
print_top_words(nmf, tfidf_feature_names, n_top_words)
```

```python
print("Extracting tf features for LDA...")
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2,
                                max_features=n_features,
                                stop_words=stop_words)
tf = tf_vectorizer.fit_transform(data_sample)
```

```python
lda = LatentDirichletAllocation(n_components=n_components, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0,n_jobs=-1)
lda.fit(tf)
```

```python
print("\nTopics in LDA model:")
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```

##### Número de topicos con respecto a su relevancia para representar la información

```python
plt.plot(s[::-1])
plt.title("#Behavior of the Topics")
```

```python
num_top_words=10

def show_topics(a):
    top_words = lambda t: [vocab[i] for i in np.argsort(t)[:-num_top_words-1:-1]]
    topic_words = ([top_words(t) for t in a])
    message=[' ,'.join(t) for t in topic_words]
    for i,j in enumerate(message):
        message1="Topic #%d: " % i
        message1+=j
        print(message1)
```

```python
print("Topic from all data")
show_topics(Vh[:10])
```

# Embedding
Se usa un modelo pre-entrenado para es español de 96 y 50 dimensiones.

```python
from pnud_textanalysis.embedding import *
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

tsne=TSNE(n_jobs=-1,n_components=2)
km = KMeans(n_clusters=5, init='k-means++', max_iter=100, n_init=1)
```

```python
sns.clustermap(A.T)
plt.title("Mapa de Clusters de Submuestra")
```

```python
#Proyección
ckm=km.fit_predict(A)
out_tsne=tsne.fit_transform(A)
```
